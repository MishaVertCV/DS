–ö—Ä–∞–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (–º–∏–Ω–∏–º—É–º –∏ –º–∞–∫—Å–∏–º—É–º) —É –±–µ—Å–ø–ª–∞—Ç–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å–º–µ—â–µ–Ω—ã –≤–ª–µ–≤–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø–ª–∞—Ç–Ω—ã—Ö –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤ - —Å—Ç–æ–∏–º–æ—Å—Ç—å –º–∏–Ω—É—Ç—ã –≤—Å–µ-—Ç–∞–∫–∏ –±–æ–ª—å—à–µ, –∑–Ω–∞—á–∏—Ç —á–∞—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–æ—Ä–æ–ø–∏—Ç—Å—è –∑–∞–≤–µ—Ä—à–∏—Ç—å –ø–æ–µ–∑–¥–∫–∏ –∑–∞ –º–µ–Ω—å—à–∏–π —Å—Ä–æ–∫. –ù–æ –ø—Ä–∏ —ç—Ç–æ–º –≤–∏–∑—É–∞–ª—å–Ω–æ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–µ —Å–∏–ª—å–Ω–æ —Å–º–µ—â–µ–Ω–æ

#### –®–∞–≥ 5. –ü–æ–¥—Å—á—ë—Ç –≤—ã—Ä—É—á–∫–∏

**5.1 –°–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –æ –ø–æ–µ–∑–¥–∫–∞—Ö –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏–∑ —à–∞–≥–∞ 4**

–°–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Å —Å—É–º–º–∞—Ä–Ω—ã–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º, –≤—Ä–µ–º–µ–Ω–µ–º, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–æ–µ–∑–¥–æ–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–π –º–µ—Å—è—Ü:

```python
new_df = df.groupby(['user_id', 'subscription_type', 'month']).agg({'distance': ['sum'], 'duration': ['sum', 'count']}).reset_index(['subscription_type'])
new_df.columns = ['subscription_type', 'distance', 'duration', 'count']
new_df.head(10)
```

**5.2 –î–æ–±–∞–≤–∏–º –≤ —ç—Ç–æ—Ç –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å—Ç–æ–ª–±–µ—Ü —Å –ø–æ–º–µ—Å—è—á–Ω–æ–π –≤—ã—Ä—É—á–∫–æ–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è**

–û–∫—Ä—É–≥–ª–∏–º –∑–Ω–∞—á–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–∏:

```python
new_df['duration'] = np.ceil(new_df['duration'])
new_df['duration'] = new_df['duration'].astype(int)
new_df.head()
```

__–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –æ—Ç —Ä–µ–≤—å—é–µ—Ä–∞ ‚Ññ1__

–û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—à—å –∏–º–µ–Ω–Ω–æ np.ceil(). –ì—Ä–∞–º–æ—Ç–Ω—ã–π —à–∞–≥

–¢–µ–ø–µ—Ä—å –ø—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ –ø–æ–º–µ—Å—è—á–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–¥–ø–∏—Å–∫–æ–π –∏ –±–µ–∑:

```python
def month_earn(row):
    sub = row['subscription_type']
    count = row['count']
    duration = row['duration']
    if sub == 'ultra':
        return (199 + duration * 6)
    else:
        return (50 * count + 8 * duration)
    

new_df['month_earnings'] = new_df.apply(month_earn, axis=1)
new_df.head()
```

–ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ç–æ–∂–µ –ø—Ä–µ–¥–ª–∞–≥–∞—é –æ–∫—Ä—É–≥–ª–∏—Ç—å:

```python
new_df['month_earnings'] = np.ceil(new_df['month_earnings'])
new_df['month_earnings'] = new_df['month_earnings'].astype(int)
new_df.head(10)
```
